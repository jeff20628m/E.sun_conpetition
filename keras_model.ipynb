# normalization : 

# formula : $$  \frac{df - df(mean)}  {df(std)} $$

```python
x_train, x_test, y_train, y_test = train_test_split( train_x_novill, train_y , train_size = 0.8, random_state =  90)
```

```python
x_train_norm = x_train
x_test_norm = x_test
```

```python
std_scale = preprocessing.StandardScaler().fit(x_train_norm)
```

```python
x_train_norm = std_scale.transform(x_train_norm)
```

```python
std_scale = preprocessing.StandardScaler().fit(x_test_norm)
x_test_norm = std_scale.transform(x_test_norm)
```

```python
model = Sequential([
    Dense(128, input_dim=476),
    Activation('relu'),
    Dropout(0.2),
    Dense(128),
    Activation('relu'),
    Dropout(0.2),
    Dense(1)
        ])

# Another way to define your optimizer
rmsprop = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0)
adam = Adam(epsilon=1e-08)
# We add metrics to get more results you want to see
model.compile(optimizer=adam,
              loss='mean_squared_error',
              metrics=['accuracy'])
```

```python
history1 = model.fit(
    x_train_norm, y_train,
 epochs= 25, batch_size = 128,
 validation_data = (x_test_norm, y_test))
```
